### Data preparation for performing  CNN-LSTM in epilepsy
### July 27, 2023

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('C://NRL/epilepsy.csv',header=None)
df.head()
```

```python
y=df[[178]]
df1=df.iloc[:,0:178]
df1.head()
```

```python
#normalize(b)
b1=df1.apply(lambda x: x/(x**2).sum()**.5, axis=1)
b1.head()
```

### Saving xtrain, xtest, ytrain, ytest for CNN-LSTM algorthm

```python
from sklearn.model_selection import train_test_split
X=b1.iloc[:,0:178]
X_train, X_test, Y_train, Y_test = train_test_split(X,y, test_size=0.1,random_state=0)
X_train=pd.DataFrame(X_train)
X_test=pd.DataFrame(X_test)
X_train.to_csv("c:/NRL/datasetsepi/train/Subs/X_train.csv",index=False,header=False)
X_test.to_csv("c://NRL/datasetsepi/test/Subs/X_test.csv",index=False,header=False)
```
